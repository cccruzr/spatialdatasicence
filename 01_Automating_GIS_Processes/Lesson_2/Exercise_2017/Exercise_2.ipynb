````python
# Determine filepath
filePath = r"data/some_posts.csv"

# Read data
df = pd.read_csv(filePath, sep=',')
df.head()
````

````python
# Create geometry series
geometry = [Point(xy) for xy in zip(df['lon'], df['lat'])]

# Drop lng, lat from df 
df = df.drop(['lon', 'lat'], axis=1)

# Define CRS
crs = {'init': 'epsg:4326'} #WGS84

# Create geodata frame 
gdf = gpd.GeoDataFrame(df, crs=crs, geometry=geometry)
gdf.head()
````

## Problem 3: Movements of individual user (optional task for advanced students)

This is an optional extra task for those who likes to learn even more. Write your codes into the same file as in previous Problem (2).

 - Group the data by userid
 - Create an empty GeoDataFrame
 - For each user: 
    - [sort](http://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.sort_values.html) the rows by timestamp 
    - create LineString objects based on the points
    - [add](http://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.append.html) the geometry and the userid into the GeoDataFrame you created in the last step
 - Determine the CRS of your GeoDataFrame to WGS84 (epsg code: 4326)
 - Save the movements of each user into a separate Shapefile
   - Name the output Shapefile based on the userid number

````python
# Sort values by 'timestamp' and then groupby 'userid'
gdf_user_group = gdf.sort_values(['timestamp']).groupby('userid')
list(gdf_user_group)
````

````python
gdf_movement = gpd.GeoDataFrame(crs=crs)
gdf_movement['geometry'] = None
#gdf_movement.loc[0,'geometry'] = Point(0, 0)
type(gdf_movement['geometry'])
````

````python
# Iterate over the group object
for key, user in gdf_user_group:
    # Create line array
    line_points = []
    
    # User ID
    user_id = int(key)
    
    # Select locations associated to individual user
    subset = user['geometry'][user['userid'] == key]
    #print(subset.shape)
    
    # If user has no more than one location: geometry is null   
    if len(user) < 2:
        for i in range(len(subset)):
            geometry = np.NaN
            #print(geometry)
    else:
        for i in range(len(user)):
            #print(user.iloc[i,2])
            line_points.append(user.iloc[i,2])
        geometry = LineString(line_points)
        #print(geometry)
        
    gdf_movement.loc[key, 'geometry'] = geometry
gdf_movement['userid']=gdf_movement.index.astype(int)
gdf_movement = gdf_movement.dropna()
````

````python
print(gdf_movement.shape)
gdf_movement.head()
````

````python
type(gdf_movement)
gdf_mov_meter = gdf_movement.to_crs(epsg=32735)
gdf_mov_meter.head()
````

````python
gdf_mov_meter['distance'] = gdf_mov_meter.length / 1000
gdf_mov_meter.head(10)
````

````python
# What was the shortest distance travelled in meters?
min_dist = gdf_mov_meter['distance'].min()

# What was the mean distance travelled in meters?
mean_dist = gdf_mov_meter['distance'].mean()

# What was the maximum distance travelled in meters?
max_dist = gdf_mov_meter['distance'].max()

print("Min distance: %.0f km \nMean distance: %.0f km \nMax distance: %.0f km" % (min_dist, mean_dist, max_dist))
````
